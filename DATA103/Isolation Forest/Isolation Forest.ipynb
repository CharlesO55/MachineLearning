# Classes

## Nodes

```python
class Node:
    def __init__(self, parent_Node, child_Values, depth, max_depth):
        '''
            Recursively create children nodes as long as child_Values can be split.
        '''
        self.parent_Node = parent_Node;
        self.child_Values = child_Values;
        self.depth = depth;
        self.max_depth = max_depth
        
        self.split_Child_Values();

    def __str__(self):
        '''
            Printing only. Can ignore.
        '''
        parent_Value = None   if self.parent_Node is None    else round(self.parent_Node.node_Value, 2);
        l_Child_Value = None  if self.L_Node is None         else round(self.L_Node.node_Value, 2);
        r_Child_Value = None  if self.R_Node is None         else round(self.R_Node.node_Value, 2);
        
        ws_char = 'x' if self.L_Node is None else '|' 
        
        whitespace = '';
        for i in range(self.depth * 5):
            if i % 5 == 0:
                whitespace += ws_char;
            else:
                whitespace += ' ';
        
        return f'{whitespace}Parent Node      : {parent_Value}\n{whitespace}-Current Node {self.depth}  : {self.node_Value:.2f}\n{whitespace}--Left Child     : {l_Child_Value}\n{whitespace}--Right Child    : {r_Child_Value}';


    def split_Child_Values(self):
        '''
        Must be a pure binary tree.
        All created nodes must either have
            1. 0 children (leaf)
            2. Exactly 2 children
        
        STOP WHEN...
            1. All the children are equal
            2. The max depth has been reached
        '''
        if (self.child_Values == self.child_Values[0]).all() or self.depth >= self.max_depth:
            # print(f'STOPPPPPPPPPPPPPPPPPPPPPPP{self.child_Values}');
            self.L_Node = None; 
            self.R_Node = None;
            self.node_Value = self.child_Values[0];
            return;

        
        threshold = np.random.uniform(min(self.child_Values) + 0.00001, max(self.child_Values));
        self.node_Value = threshold;
        
        split_index = np.where(self.child_Values > threshold)[0][0];


        # print(f'\n\tOriginal: {self.child_Values}')        
        # print(f'Threshold: {threshold:.2f}');
        # print(f'Split by : {self.child_Values[:split_index]} :: {self.child_Values[split_index]} :: {self.child_Values[split_index:]}')
        
        self.L_Node = self.create_Node(self.child_Values[:split_index]);
        self.R_Node = self.create_Node(self.child_Values[split_index:]);


    def create_Node(self, values):
        return Node(self, values, self.depth + 1, self.max_depth)
```

## Trees

```python
class Tree:
    def __init__(self, values, attr = None, max_depth = 20):
        '''
            Start creating the tree from the root node. 
            Sort first for easier splitting.
        '''
        values = np.array(sorted(values))

        self.attr = attr;
        self.root = Node(None, values, 0, max_depth)

        self.deepest_depth = 0
        
        self.tree_traversal(self.root)
    
    def tree_traversal(self, node, print_Debug=False):
        '''
            Preorder tree traversal.
        '''
        if node is None:
            return;

        if print_Debug:
            print(node)

        self.deepest_depth = max(node.depth, self.deepest_depth)
        
        self.tree_traversal(node.L_Node)
        self.tree_traversal(node.R_Node)


    def find_value(self, value, print_Debug = True):
        '''
            Find how deep/number of nodes needed to traverse before finding the value.
            Return the `depth`
        '''
        curr_Node = self.root
        traversed_Nodes = []
        
        while curr_Node is not None:
            traversed_Nodes.append(curr_Node)
            
            if value <= curr_Node.node_Value:
                curr_Node = curr_Node.L_Node
            else:
                curr_Node = curr_Node.R_Node

        if print_Debug:
            for node in traversed_Nodes:
                print(node)
    
        return traversed_Nodes[-1].depth
```

# Isolation Forests

Isolation forests are a type of decision tree used in anomaly detection. Aside from generating label predictions, it's also able to provide an anomaly_score 


It works by using decision trees to split random features according to a random threshold (within the values limits).

## Functions required
#### `fit` : Train the model

#### `predict` : Score classifying as an anomaly (-1) or normal (1).


```python
def calc_Leaf_Impurity(val1, val2):
    total = val1 + val2
    return 1 - np.square(val1 / total) - np.square(val2 / total);
```

```python
print(calc_Leaf_Impurity(3, 1))

print(calc_Leaf_Impurity(1, 2))
```

```python
def calc_Total_Impurity(values):
    
    total_impurity = 0;
    for val in values:
        total_impurity += calc_Leaf_Impurity(val[0], val[1]) *  (np.sum(val)/ np.sum(values)) 

    return total_impurity;

calc_Total_Impurity([[3,1], [1,2]])
```

```python
import math

class iForest:
    def __init__(self, n_trees=100, n_subsampling_size=256):
        self.n_trees = n_trees
        self.n_subsampling_size = n_subsampling_size

        # self.max_depth = math.ceil(math.log2(n_subsampling_size))
        self.max_depth = 100
        self.trees = []

    def fit(self, X):
        for t in range(self.n_trees):
            # SELECT A RANDOM ATTRIBUTE
            attr = np.random.choice(X.columns)
            # print(t, ': ', attr);

            # BUILD A TREE USING A RANDOM SUBSET OF THE RANDOM ATTRIBUTE
            newTree = Tree(self.do_subsampling(X[attr]), attr, self.max_depth);

            # ATTACH THE TREE TO THE FOREST
            self.trees.append(newTree)
            # newTree.tree_traversal(newTree.root)
        
    def do_subsampling(self, X):
        '''
            "Throughout this paper, sub-sampling is conducted by random selection of instances without replacement"
        
            Smaller sample sizes makes it easier to spot outliers as this reduces the effects of masking and swamping.
            Difficulties of isolating points from clusters (from making more cuts) are thereby minimized.
        '''
        rng = np.random.choice(X.index, replace=False, size=self.n_subsampling_size);
        return X.loc[rng]
        

    def predict(self, x_test):
        cum_depth = 0
        cum_max_depth = 0
        for t in self.trees:
            score =  t.find_value(x_test[t.attr], False)

            print(f'{t.attr[:6]}:\t {score}/{t.deepest_depth} = {score/t.deepest_depth}')
            
            cum_depth += score;
            cum_max_depth += t.deepest_depth
        print(f'Total depth {cum_depth}')
        print(f'Avg depth {cum_depth/cum_max_depth}')
            
forest = iForest(5)
# forest.fit(X[['age', 'bmi', 'avg_glucose_level']])
forest.fit(X)
forest.predict(X.iloc[40])
```
